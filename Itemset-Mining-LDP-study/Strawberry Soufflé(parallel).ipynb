```python
def laplace_mech(answer, sensitivity, epsilon):
    return answer + np.random.laplace(loc=0, scale=sensitivity / epsilon)
```

### LDP

#### DE

```python
def DE_p_q(epsilon, d):
    p = (e**epsilon) / (e**epsilon + d - 1)

    q = (1 - p) / (d - 1)

    return p, q
```

```python
#case de mehcanism decides the response to be fake, ramdomly picks all the values but the truthful response of the domain
def random_pick_DE(domain, response):
    domain_q = [i for i in domain if i != response]
    return random.choice(domain_q)
```

```python
def randomized_responses_DE(responses, epsilon):
    d = len(responses)
    domain = [i for i in range(0, d)]
    p, q = DE_p_q(epsilon, d)
    #print("responses: ",responses)
    noisy_responses = np.zeros(len(responses))
    
    for i in range(len(responses)):
        #print("i:",i)
        for j in range(responses[i]):
            #print("j:",j)
            sample = np.random.random()

            if sample <= p:
                #tell the truth
                #by adding to noisy_responses in the i position
                
                noisy_responses[i] = noisy_responses[i] + 1
                #print("truth: ", noisy_responses)
    
            else:
                #lie
                #by adding to noisy_responses in any position but i 
                any_pos_but_i = random_pick_DE(domain, i)
                #print("lie: ",any_pos_but_i)
                noisy_responses[any_pos_but_i] = noisy_responses[any_pos_but_i] + 1
                #print("noisy_responses added: ", noisy_responses)
            
                
    return noisy_responses
```

```python
def adjustDE(noisy_responses, epsilon, n):
    d = len(noisy_responses)
    #print("Deresponses1  d: ",d)
    p, q = DE_p_q(epsilon, d)
    #print("Deresponses  p,q: ",p,q)
    copy = noisy_responses.copy()
    #print("copy ",copy)
    #print(sum(copy))

    a = 1.0 / (p - q)
    b = n * q / (p - q)
    
    for i in range(len(copy)):
        copy[i] = a * copy[i] - b
        
    #print("copy adjusted: ",copy)
    #print(sum(copy))
    #print(copy)
    return copy
```

```python
def DEresponses(responses, epsilon):
    n = sum(responses)
    #print("Deresponses  n: ",n)
    return n_sub(n, adjustDE(randomized_responses_DE(responses, epsilon), epsilon, n))
```

#### OUE

```python
def optimized_p_q(epsilon):
    p = 0.5
    q = 1/(e**epsilon + 1)

    return p, q
```

```python
def OUEresponses(responses, epsilon):
    p, q = optimized_p_q(epsilon)
    n = sum(responses)
    return n_sub(n, adjustUE(perturb_onto_agg(responses, p, q), p, q, n))  
```

#### UE

```python
def unary_epsilon(p, q):
    return np.log(p*(1-q)/((1-p)*q))
```

```python
def symetric_p_q(epsilon):
    p = (e**(epsilon/2)) / (e**(epsilon/2) + 1)
    q = 1 / (e**(epsilon/2) + 1)

    return p, q
```

```python
def perturb_onto_agg(responses, p , q):
    noisy_responses = np.zeros(len(responses))
    #print("SOMA: ",sum(responses))
    
    for i in range(len(responses)):
        #print("i",i)
        for j in range(responses[i]):
            #print("j",j)
            for k in range(len(noisy_responses)):
                #print("k",k)
                #works hence the internal for iterates each one time for a singular unit
                sample = np.random.random()
                
                #case bit is 1(same cell, therefore using p)
                if i == k:
                    #print("here")
                    if sample <= p:
                        #tell the turht
                        #by adding 1 to the i(k) position of the noisy vector
                        noisy_responses[k] = noisy_responses[k] + 1
                    #else:
                        #lie
                        #by not adding anything to the vector
                #handles other cells(using q probability)
                else:
                    #print("there")
                    if sample <= q:
                        #lie
                        #by summing 1 onto the k position of the noisy vector
                        noisy_responses[k] = noisy_responses[k] + 1
                    #else: 
                        #truth
                        #by omiting form it
    return noisy_responses
```

```python
def adjustUE(noisy_responses, p, q, n):
    d = len(noisy_responses)
    copyUE = noisy_responses.copy()

    a = 1.0 / (p - q)
    b = n * q / (p - q)
    
    for i in range(len(copyUE)):
        copyUE[i] = a * copyUE[i] - b

    return copyUE
```

```python
def UEresponses(responses, epsilon):
    p, q = symetric_p_q(epsilon)
    n = sum(responses)
    return n_sub(n, adjustUE(perturb_onto_agg(responses, p, q), p ,q, n))
```

## Preprocessing

```python
#use this one
def n_sub(n ,est_dist):
    est_dist = est_dist.astype(float)
    #print("naa = ", n)
    #print(est_dist)
    #print(sum(est_dist))
    while (np.fabs(sum(est_dist) -n) > 1) or (est_dist < 0).any():
        est_dist[est_dist < 0] = 0
        #print("distribuição: ", est_dist)
        total = sum(est_dist)
        #print("soma: ", total)
        mask = est_dist > 0
        #print("máscara: ", mask)
        diff = (n - total)/sum(mask)
        #print("diff = ", diff)
        est_dist[mask] += diff
    #print(est_dist)
    #x = est_dist.round().astype(int)
    #print(x)
    #print(sum(x))
    return est_dist.round().astype(int)
```

## PSFO

```python
def generate_dummy_pool(l, I):
    max_value = max(I) + l
    domain_set = set(range(1, max_value + 1))
    
    original_set = set(I)
    dummy_set = list(domain_set - original_set)

    if len(dummy_set) >= l:
        dummy_set = random.sample(dummy_set, l)

    return dummy_set
```

```python
# padding and sampling function
 #case cardinality of v is less than l adds l - |transaction| (dummy variables), dont belong to the items of transaction, so get random ones form the domain that dont belong to the transaction

 #then sample! That is: select AN element of trnasaction at a uniform random

#padds and samples, ,lol
def PS(l, transaction, pool):

    if len(transaction) < l:
        while len(transaction) < l:
            dummy = random.choice(pool)
            transaction.append(dummy)  #adds a dummy item from the domain pool

    sample = random.choice(transaction)
    
    return sample
```

```python
#Applies padding and sampling then an FO protocol
#To estimate the frequency of an item
#x, one obtains the frequency estimation of x from the FO
#protocol, and then multiplies it by l

#alsdo uses adaptive FO based on l and d

def PSFO(l, transactions, epsilon, new_domain, pool):
        
    #ground truth for preprocessing
    #sum of real distribution cna be attained using DP
        
    #new domain which the FO works with
    #print("New domain: ", new_domain)
    d = len(new_domain)
    #print("d: ", d)
    r = np.zeros(d, dtype=int)
    
    #generates a list of size n, that is, n values, each coresponding to a padded and sampled transaction
    #for each transactions applies PS
    for transaction in transactions:
        #print("transaction",transaction)
        sample = PS(l, transaction, pool)
        #print("sample: ",sample)

        r += encode(sample, new_domain)
        #print(r)
        
    #print("r :", r)
    
    #adaptive FO check with new epsilon for DE
    #print(d, " <>",   (l*(4*l - 1)*e**(epsilon) + 1))
    if d < (l*(4*l - 1)*e**(epsilon) + 1):
        new_epsilon = np.log(l*(e**epsilon -1)) + 1
        #print("DE e:", new_epsilon)
        r_noise = DEresponses(r, new_epsilon)
    else:
        r_noise = OUEresponses(r, epsilon)
        #print("OUE e:", epsilon)
    #print(sum(r_noise))
    #print("r_noise: ", r_noise)
    #estimations times l
    est = [item * l for item in r_noise]
    
    return est 
```

## Cardinalities Reconstruction

```python
def get_intersections_cardinalities(S, transactions, epsilon):
    
    intersection_cardinalities = []
    r = np.zeros(len(S) + 1, dtype=int)

    #intersection
    for transaction in transactions:
        intersection_cardinality = sum(item in S for item in transaction)
        intersection_cardinalities.append(intersection_cardinality)

    domain = list(range(0, len(S)+1))
    
    for cardinality in intersection_cardinalities:
        r += encode(cardinality, domain)

    #OUE used by default
    r_noise = OUEresponses(r, epsilon)
    #print(intersection_cardinalities)
    #print(sum(intersection_cardinalities))
    return r_noise
```

```python
def reconstruct_cardinalities(r):
    return np.hstack([np.full(count, idx) for idx, count in enumerate(r)])
```

## Partitioning(groups)

```python
def Partition_groups(user_transactions, p, percentage_1, percentage_2, percentage_3, percentage_4, percentage_5):
    total_users = len(user_transactions)
    
    item_group_size = int(total_users * p) #reserved group size for item estimations
    itemset_group_size = int(total_users - item_group_size)  #reserved group size for itemset estimations
    
    group_1_size = int(item_group_size * percentage_1)  #for candidates; S
    group_2_size = int(item_group_size * percentage_2)  #for; L
    group_3_size = int(item_group_size - (group_1_size + group_2_size) )  #estimation of S

    group_4_size = int(itemset_group_size * percentage_4)  #
    group_5_size = int(itemset_group_size - group_4_size)  #
    
    #shuffle list
    random.seed(42)  # For reproducibility
    shuffled_data = user_transactions.copy()
    random.shuffle(shuffled_data)
    
    #groups(items)
    group_1 = shuffled_data[:group_1_size]  
    group_2 = shuffled_data[group_1_size:group_1_size + group_2_size]  
    group_3 = shuffled_data[group_1_size + group_2_size:group_1_size + group_2_size + group_3_size]  

    #groups(itmesets)
    group_4 = shuffled_data[group_1_size + group_2_size + group_3_size:group_1_size + group_2_size + group_3_size + group_4_size]
    group_5 = shuffled_data[group_1_size + group_2_size + group_3_size + group_4_size:]
    #print(len(group_1))
    #print(len(group_2))
    #print(len(group_3))
    #print(len(group_4))
    #print(len(group_5))
    return group_1, group_2, group_3, group_4, group_5 
```

```python
def get_domain(user_transactions):
    return list(set(item for transaction in user_transactions for item in transaction))
```

## Prune the domain (Step 1)

```python
def Candidates_SVIM(z, k, transactions, epsilon, domain):
    
    l = 1

    #gets the new domain wiht the pool of dummy items
    pool = generate_dummy_pool(l, domain)
    new_domain = domain.copy()
    new_domain.extend(pool)
    new_domain.sort()

    r = PSFO(l, transactions, epsilon, new_domain, pool)

    #top values 
    top_2k_values = np.sort(np.unique(r))[-z*k:][::-1]
    top_2k_indices = [np.where(r == value)[0] for value in top_2k_values]
    
    S = [int(arr[0]) if not isinstance(arr, int) else arr for arr in top_2k_indices]

    return S
```

## Size Estimation (Step 2)

```python
#some estimates may be overwhelmed by noise
def apply_threshold(estimations, zk, epsilon, n, alpha = 0.05):

    significance_lvl = 1 - (alpha/(zk))
    #F-1 , the inverse of standart normal CDF
    z_score = norm.ppf(significance_lvl)

    #Variance of the FO (OUE)
    Var = (n*4*e**(epsilon)) / (e**epsilon - 1)**2

    T = z_score*np.sqrt(Var)

    thresholded_estimations = [est if est >= T else 0 for est in estimations]

    return thresholded_estimations
```

```python
def L_SVIM(S, percentile, transactions, epsilon):

    r = get_intersections_cardinalities(S, transactions, epsilon)
    
    #cahnge this L_SVIM now gets k and n
    thresholded_r = apply_threshold(r, len(S), epsilon, sum(r))
    
    #estimated_cardinalities = reconstruct_cardinalities(r)
    
    thres_estimated_cardinalities = reconstruct_cardinalities(thresholded_r)

    #L = np.percentile(estimated_cardinalities, percentile)

    if len(thres_estimated_cardinalities) > 0:
        L_new = np.percentile(thres_estimated_cardinalities, percentile)
        if L_new == 0:
            L_new = 1
    else:
        L_new = 1

    return int(L_new), thresholded_r
```

## Candidate Estimation (Step 3)

```python
def Candidates_est_SVIM(S, L, transactions, epsilon):

    intersections = []

    for transaction in transactions:
        intersection = list(set(transaction).intersection(S))
        
        #saves all the intersection
        #virtually replacing the transactions list of list to a
        #list of intersections for each transaction with S
        intersections.append(intersection)    

    domain = S

    pool = generate_dummy_pool(L, domain)
    new_domain = domain.copy()
    new_domain.extend(pool)
    new_domain.sort()
    #print(new_domain)
    #print(len(new_domain))
    #print(pool)

    #print(S)
    r = PSFO(L, intersections, epsilon, new_domain, pool)
    #print(r)
    filtered_r = [r[new_domain.index(s)] for s in S if s in new_domain]
    
    return filtered_r
```

## Estimation Update (Step 4)

```python
def Estimation_update_SVIM(L, intersection_frequencies):
    term_1 = 0
    term_2 = 0
    for i in range(len(intersection_frequencies)):
        #l times summation of distribution
        #number of items
        term_1 += i*intersection_frequencies[i] 

        #umber of missed items
        if i >= (L + 1):
            term_2 += (i-L)*(intersection_frequencies[i])

    if term_1 - term_2 == 0:
        update_factor = 1
    else:
        update_factor = term_1 / (term_1 - term_2)
    
    return update_factor
```

## SVIM vanilla

```python
def SVIM(group_1, group_2, group_3, z, k, epsilon, domain):

    #step 1(finds candidates)
    S = Candidates_SVIM(z, k, group_1, epsilon, domain)
    #print("\033[1mCandidates:\033[0m ", S)

    #step 2(estimates size)
    percentile = 90
    L_new, r_thresholded = L_SVIM(S, percentile, group_2, epsilon)
    #print("\033[1mEstimated Size(L):\033[0m ", L_new)

    #step 3(estimates frequencies of said candidates)
    est_frequencies = Candidates_est_SVIM(S, L_new, group_3, epsilon)
    #print("\033[1mEstimated Frequencies:\033[0m ", est_frequencies)

    #step 4(estimation update)
    update_factor = Estimation_update_SVIM(L_new, r_thresholded)
    #print("\033[1mUpdate Factor:\033[0m ", update_factor)

    #update
    est_frequencies_updated = [int(item * update_factor) for item in est_frequencies]
    #print("\033[1mUpdated Estimated Frequencies:\033[0m ", est_frequencies_updated)

    ## Rearange order according to new estimations and return 2 lists, the items and their frequencies accondinly
    # Zip the lists together and sort by the values in b (in descending order)
    #zipped_lists = sorted(zip(est_frequencies_updated, S), reverse=True)
    
    #items_frequencies, frequent_items = zip(*zipped_lists)
    
    #items_frequencies = list(items_frequencies)
    #frequent_items = list(frequent_items)
    
    #print("\033[1mSorted a:\033[0m ", items_frequencies)
    #print("\033[1mSorted b:\033[0m ", frequent_items)

    return S[:k], est_frequencies_updated[:k]
```

```python
def ground_truth(user_transactions, k):
    flattened_list = [item for sublist in user_transactions for item in sublist]

    item_counts = Counter(flattened_list)
    
    top_k_items = item_counts.most_common(k)
    
    # Separate items and frequencies into two vectors (lists)
    items = [item for item, _ in top_k_items]
    frequencies = [frequency for _, frequency in top_k_items]

    return items, frequencies
```

# Metrics (NCR and Var)

## NCR

```python
def create_quality_function(xt, k):
    # Generate scores based on k
    scores = [max(k - i, 0) for i in range(len(xt))]
    
    # Create a mapping function
    def quality_function(value):
        if value in xt:
            index = xt.index(value)
            return scores[index]
        else:
            return 0  # Return 0 if the value is not in xt
    
    return quality_function
```

```python
def NCR(xr, xt, k):
    
    qr_sum = 0
    qt_sum = 0

    quality_function = create_quality_function(xt, k)
    
    for item in xr:
        qr_sum += quality_function(item)
        #print("item", item)
        #print("value", quality_function(item))

    #print("-------------------------")
    for item in xt:
        qt_sum += quality_function(item)
        #print("item", item)
        #print("value", quality_function(item))

    qr_sum = qr_sum/( k*(k+1)/2)
    qt_sum = qt_sum/( k*(k+1)/2)
    
    NCG = qr_sum/qt_sum

    return NCG
```

## Variance

```python
# n is the number of users, thereby transaction
# fx is the frequency the itemset

def Var(xr, xt, candidates_estimation, real_frequencies, n):

    succ_identified = ( list(set(xr).intersection(xt)) )
    #print(succ_identified)
    summation = 0
    
    for item in succ_identified:
        #for the items in the intersection(aka the heavy hitter sucesfully identified)
        #calculate the difference between the actual frequency and the estimated one
        #square it up
        #print(item)
        xr_index = xr.index(item)
        #print(xr_index)
        xt_index = xt.index(item)
        #print(xt_index)
        # Retrieve the corresponding frequencies
        fx = real_frequencies[xt_index]  # Frequency from the real frequencies list
        phi = candidates_estimation[xr_index]  # Estimated frequency
        summation += (fx - phi)**2

    return (1/len(succ_identified))*summation if succ_identified else 0
```

```python
# does the experiment n times
def experiment(group_1, group_2, group_3, z, k, epsilon, domain, times):
    ncr_results = []
    var_results = []

    n = len(group_3)

    for i in range(times):
        #ground truth
        St, Fx = ground_truth(group_1, k)
        Sx, Ft = ground_truth(group_3, k)
        
        #estimation
        S, F = SVIM(group_1, group_2, group_3, z, k, epsilon, domain)
    
        ncr = NCR(S, St, k)
        var = Var(S, St, F, Ft, n)
    
        ncr_results.append(ncr)
        var_results.append(var)
    
    return ncr_results, var_results 
```

```python
def multi_plotter(data, y_metric, x_metric, y_lists, y_errors, x_values, z_term, z_values, fixed_term, fixed_value):
    # Apply seaborn ticks style for scientific paper aesthetics
    sns.set(style="ticks", palette="muted")  # Use 'ticks' style

    # Check if log scale is needed for the y-axis
    log_scale = y_metric.lower() == 'var'

    # Set up colors and markers
    colors = plt.cm.viridis(np.linspace(0, 1, len(y_lists)))
    markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h', 'H', 'x', '+']

    plt.figure(figsize=(12, 6))

    for idx, (y_values, y_err, color) in enumerate(zip(y_lists, y_errors, colors)):
        marker_style = markers[idx % len(markers)]
        plt.errorbar(
            x_values, y_values, yerr=y_err, marker=marker_style, markerfacecolor='none', alpha=0.8,
            color=color, label=f'{z_term} {z_values[idx]}',
            capsize=4, elinewidth=1.5, markeredgewidth=1.5, ms = 10
        )

    # Titles and labels with larger font sizes
    plt.title(f'{data}, {y_metric}, vary {x_metric}, {fixed_term} = {fixed_value}', fontsize=16)
    plt.xlabel(x_metric, fontsize=14)
    plt.ylabel(y_metric, fontsize=14)

    if log_scale:
        plt.yscale('log')

    # Enhance legend and grid
    plt.legend(fontsize=12)
    plt.grid(visible=True, which='both', axis='y', linestyle=':', linewidth=1.4)
    plt.grid(visible=True, which='both', axis='x', linestyle=':', linewidth=1.4)

    # Customize spines for a boxed appearance
    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)

    # Make ticks larger and bolder
    ax.tick_params(axis='both', which='major', labelsize=12, width=1.5, length=8, direction='in')
    ax.tick_params(axis='both', which='minor', labelsize=10, width=1, length=8, direction='in')

    # Ensure ticks are present on all sides of the plot
    ax.xaxis.set_ticks_position('both')
    ax.yaxis.set_ticks_position('both')

    plt.tight_layout()
    plt.show()
```

```python
def comparator(start, final, pace, group_1, group_2, group_3, k, domain, times):
    z = 2
    epsilon_values = list(np.arange(0, final + pace, pace))
    epsilon_values[0] = start

    ncr_values = []
    var_values = []
    ncr_errors = []
    var_errors = []
    
    for i in epsilon_values:
        epsilon = i

        #returns the results for each experiemn
        #do a mean after and then calculate the errors bars
        ncr_results, var_results = experiment(group_1, group_2, group_3, z, k, epsilon, domain, times)

        mean_ncr = sum(ncr_results)/times
        mean_var = sum(var_results)/times
        error_ncr = np.std(ncr_results) / np.sqrt(times)
        error_ncr = np.std(ncr_results) / np.sqrt(times)

        #means
        ncr_values.append(mean_ncr)
        var_values.append(mean_var)

        #errors to mean
        ncr_errors.append(error_ncr)
        var_errors.append(error_var)
        
    return ncr_values, var_values, ncr_errors, var_errors
```

## Test 1 (varying the partitioning of the groups)

```python
#testing group partitionning 
def comparator_partitioning(start, final, pace, pace_prctg, user_transactions, k, times):
    z = 2
    epsilon_values = list(np.arange(0, final + pace, pace))
    epsilon_values[0] = start
    
    ncr_lists = []
    var_lists = []
    ncr_errors = []
    var_errors = []

    domain = get_domain(user_transactions)

    percentages_1 = [round(num, 1) for num in list(np.arange(0.10, 0.90 + pace_prctg, pace_prctg))]
    percentages_1 = np.clip(percentages_1, 0.1, 0.9).tolist()

    #for each partition calculate, using the range of budgets, the results, end iwll ahve a alsit of lsits, eahc element of the lsits being the corresponding partition results
    for j in percentages_1:
        #calc percentages of other groups
        percentage_2 = round((1 - j)*0.2, 2)
        percentage_3 = round((1 - j)*0.8, 2)
        
        #gets partitioning
        group_1, group_2, group_3,_,_ = Partition_groups(user_transactions, 0.50, j, percentage_2, percentage_3, 0.1, 0.9)
        
        temp_ncr_values = []
        temp_var_values = []
        temp_ncr_errors = []
        temp_var_errors = []
        
        for i in epsilon_values:
            epsilon = i

            #returns the results for each experiemn
            #do a mean after and then calculate the errors bars
            ncr_results, var_results = experiment(group_1, group_2, group_3, z, k, epsilon, domain)

            mean_ncr = sum(ncr_results)/times
            mean_var = sum(var_results)/times
            error_ncr = np.std(ncr_results) / np.sqrt(times)
            error_ncr = np.std(ncr_results) / np.sqrt(times)

            #means
            temp_ncr_values.append(mean_ncr)
            temp_var_values.append(mean_var)

            #errors to mean
            temp_ncr_errors.append(error_ncr)
            temp_var_errors.append(error_var)
            

        ncr_lists.append(temp_ncr_values)
        var_lists.append(temp_var_values)
        
        ncr_errors.append(temp_ncr_errors)
        var_errors.append(temp_var_errors)

    return ncr_errors, var_errors, ncr_lists, var_lists
```

## Test 2 (varying factor of k)

```python
#testing group partitionning 
def comparator_constant_factor(start, final, pace, pace_factor, end_factor, user_transactions, k, times):

    epsilon_values = list(np.arange(0, final + pace, pace))
    epsilon_values[0] = start
    
    ncr_lists = []
    var_lists = []

    domain = get_domain(user_transactions)

    #default partitioning
    group_1, group_2, group_3,_,_ = Partition_groups(user_transactions, 0.50, 0.5, 0.1, 0.4, 0.1, 0.9)

    factors = [round(num, 1) for num in list(np.arange(1, end_factor + pace_factor, pace_factor))]

    #for each partition calculate, using the range of budgets, the results, end iwll ahve a alsit of lsits, eahc element of the lsits being the corresponding partition results
    for j in factors:
        
        temp_ncr_values = []
        temp_var_values = []
        
        for i in epsilon_values:
            epsilon = i

            mean_ncr, mean_var = experiment(group_1, group_2, group_3, j, k, epsilon, domain, times)
            
            temp_ncr_values.append(mean_ncr)
            temp_var_values.append(mean_var)

        ncr_lists.append(temp_ncr_values)
        var_lists.append(temp_var_values)

    colors = plt.cm.plasma(np.linspace(0, 1, len(ncr_lists)))

    # Plot NCR values
    plt.figure(figsize=(24, 8))
    for idx, (ncr_values, color) in enumerate(zip(ncr_lists, colors)):
        plt.plot(epsilon_values, ncr_values, marker='x', alpha=0.7, color=color, label=f'Factor of k {factors[idx]}')
    
    plt.title('POS, NCR, vary epsilon, while varying k')
    plt.xlabel('epsilon')
    plt.ylabel('ncr')
    plt.legend()

    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.minorticks_on()
    
    plt.grid(axis='y', linestyle='-.', linewidth=0.5, alpha=0.5, color='black')
    
    plt.tight_layout()
    plt.show()

    # Plot Var values
    plt.figure(figsize=(24, 8))
    for idx, (var_values, color) in enumerate(zip(var_lists, colors)):
        plt.plot(epsilon_values, var_values, marker='v', alpha=0.7, color=color, label=f'Factor of k {factors[idx]}')
    
    plt.title('POS, Var, vary epsilon, while varying k')
    plt.xlabel('epsilon')
    plt.ylabel('var')
    plt.yscale('log')
    plt.legend()

    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.minorticks_on()
    
    plt.grid(axis='y', linestyle='-.', linewidth=0.5, alpha=0.5, color='black')
    
    plt.tight_layout()
    plt.show()
```
